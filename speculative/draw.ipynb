```python
from vllm import LLM, SamplingParams
import __init__

prompts = [
    "The future of AI is what",
    "The future of AI is what",
]
sampling_params = SamplingParams(temperature=0.8, top_p=0.95, max_tokens=1000)

llm = LLM(
    model="deepseek-ai/DeepSeek-R1-Distill-Qwen-32B",
    tensor_parallel_size=2,
    speculative_model="deepseek-ai/DeepSeek-R1-Distill-Qwen-7B",
    num_speculative_tokens=5,
    use_v2_block_manager=True,
)
outputs = llm.generate(prompts, sampling_params)

for output in outputs:
    prompt = output.prompt
    generated_text = output.outputs[0].text
    print(f"Prompt: {prompt!r}, Generated text: {generated_text!r}")
```

```python
prompts = [
    "The future of AI is what I think",
]
sampling_params = SamplingParams(temperature=0.8, top_p=0.95, max_tokens=500)
outputs = llm.generate(prompts, sampling_params)
```

```python
outputs[1].outputs[0].text, len(outputs[1].outputs[0].token_ids), outputs[1].metrics.spec_token_acceptance_counts
    # spec_token_acceptance_counts: number of accepted speculative tokens at
    #                               each position; the first token is from 
    #                               the target model and is always accepted;
    #                               e.g., when it's [10, 8, 4, 2] for a req, 
    #                               it means there were 10 forward passes in
    #                               total, and there were 8, 4, 2 accepted 
    #                               tokens at 1st, 2nd, 3rd speculation step. 
```

```python
acceptance_counts = [0] * (5 + 1)
for output in outputs:
    for step, count in enumerate(output.metrics.spec_token_acceptance_counts):
        acceptance_counts[step] += count

print(f"mean acceptance length: \
    {sum(acceptance_counts) / acceptance_counts[0]:.2f}")
```

```python
import matplotlib.pyplot as plt

dataset = 'amc'
# 准备数据
labels = ['1.5B', '1.5B+32B', '32B']

# 学术风格颜色（ColorBrewer 配色）
colors = ['#66c2a5', '#fc8d62', '#8da0cb']
acc_values = [round(i*100,2) for i in data[dataset]['acc_values']]
token_values = data[dataset]['token_values']

# 设置图像大小与分辨率
plt.figure(figsize=(6, 2), dpi=300)

# ----------- 图1: Accuracy -----------
plt.subplot(1, 2, 1)
bars1 = plt.bar(labels, acc_values, color=colors, edgecolor='black')
for bar in bars1:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval + 0.005,
             f"{yval:.2f}", ha='center', va='bottom', fontsize=12)
if dataset == 'math500': plt.ylim(80, 95)
if dataset == 'aime': plt.ylim(20, 75)
if dataset == 'gpqa': plt.ylim(30, 68)
if dataset == 'amc': plt.ylim(70, 100)
plt.ylabel('Accuracy', fontsize=11)
# plt.title('Accuracy Comparison', fontsize=13)
plt.xticks(rotation=0, fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.5)

# ----------- 图2: Avg Token Num -----------
plt.subplot(1, 2, 2)
bars2 = plt.bar(labels, token_values, color=colors, edgecolor='black')
for bar in bars2:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval + 50,
             f"{yval:.0f}", ha='center', va='bottom', fontsize=12)
plt.ylabel('Average Length', fontsize=11)
if dataset == 'math500':  plt.ylim(3500, 5900)
if dataset == 'aime': plt.ylim(12000, 19000)
if dataset == 'gpqa': plt.ylim(5200, 8500)
if dataset == 'amc': plt.ylim(6800, 11500)
# plt.title('Average Token Number Comparison', fontsize=13)
plt.xticks(rotation=0, fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.5)

# 调整布局并保存
plt.tight_layout()
plt.savefig(f'{dataset}.pdf', dpi=300)
plt.show()
```
